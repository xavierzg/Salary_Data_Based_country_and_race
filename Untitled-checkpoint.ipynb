```python
def null_values(data,lista):
    null_val = []
    for item in lista:
        null_val.append(pd.isnull(data[item]).values.ravel().sum())
    return null_val
lista = np.array(['Unnamed: 0', 'Age', 'Gender', 'Education Level', 'Job Title',
       'Years of Experience', 'Salary', 'Country', 'Race'])
xd = null_values(data,lista)
```

```python

def change_values(data1):
    columnas = []
    for item in data1.columns.values:
        if data1[item].dtypes == 'object':
            columnas.append(item)
            
    for item in columnas:
        x = [i for i in range(pd.get_dummies(data1[item]).columns.values.shape[0])]
        data1[item] = data1[item].replace(pd.get_dummies(data1[item]).columns.values, x)
    return data1 
```

```python
change_values(data1)
```

```python
data2 = data1 
data3 = data2.dropna(axis=0,how="any")
data4 = data3.drop('Unnamed: 0', axis = 1)
data4
```

```python
import seaborn as sns
def matrix_corr(data): #matriz de correlación
    correlation = data.corr()
    sns.heatmap(correlation, annot=True, cbar=True, cmap="RdYlGn")

matrix_corr(data4)
```
